1. (1p) Se dă o listă de numere. Să se retrurneze lista numerelor pare, în ordine inversă a aparițiilor.

Exemplu: [23, 45, 4, 658, 8, 9, -4] => [-4, 8, 658, 4]

```python
lista1 = [23, 45, 4, 658, 8, 9, -4]
# variabila rezultat va fi  rezultat_lista_pare
rezultat_lista_pare=[]
for i in reversed(lista1):
    if i % 2 == 0:
        rezultat_lista_pare.append(i)
#         print(i);print('')
print("Lista rezultat este: "); print(rezultat_lista_pare)

# assert rezultat_lista_pare == [-4, 8, 658, 4], "Rezultat asteptat: [-4, 8, 658, 4]"
```

7. (1p) Să se verifice dacă o propoziție dată este palindrom sau nu (fără a lua în considerare semnele de punctuație sau diacriticele).

Exemple:

    1. ELE FAC CAFELE.
    2. ALA E SERGIU, NU-I GRESEALA.
    3. AI RAMAS ACASA, MARIA?

```python
def isPalindrome(string):
    return string == string[::-1]
string1 = "ELE FAC CAFELE."
string2 = "ALA E SERGIU, NU-I GRESEALA."
string3 = "AI RAMAS ACASA, MARIA?"

string_1 = []
string_2 = []
string_3 = []

for character in string1:
    if character.isalpha():
        string_1.append(character)
    
for character in string2:
    if character.isalpha():
        string_2.append(character)
    
for character in string3:
    if character.isalpha():
        string_3.append(character)
    
    
valid = isPalindrome(string_1)
if valid:
    print("Sirul 1 este palindrom.")
else:
    print("Sirul 1 nu este palindrom.")
    
valid = isPalindrome(string_2)
if valid:
    print("Sirul 2 este palindrom.")
else:
    print("Sirul 2 nu este palindrom.")
    
valid = isPalindrome(string_3)
if valid:
    print("Sirul 3 este palindrom.")
else:
    print("Sirul 3 nu este palindrom.")
```
