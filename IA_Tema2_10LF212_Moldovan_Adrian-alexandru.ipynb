 (1p) 1. Să se scrie o funcție care primește ca parametru un număr de 5 cifre. Să se returneze o listă cu toate numerele prime formate din cifrele parametrului.

```python
from itertools import permutations

fiveDigitsNumber = 13579

def isPrime(nr:int) -> bool:
    '''This function shows if a integer number is prime or not.
    param nr: The number that will be searched if prime or not.
    for and if: Classic isPrime algorithm.
    '''
    for index in range(2,nr):
        if (nr%index) == 0:
              return False
    return True
    
def allPermutationList(number:int) -> str:
    '''This function will take all permutations of a 5 digits number,
    we will transform every permutation into a string of digits and put it in a list.
    Now, for every permutation we will construct a int number and if the number isPrime we add that permutation
    to a primeNumberList. At the end we print all the permutations from that list.
       param number: The 5 digits number from we make the permutations.  
    '''
    primeNumberList = []
    digitsList = []
    for index in str(number):
        digitsList.append(index)
    all_permutations = permutations(digitsList)
    for permutation in all_permutations:
        number_construction = 0
        for char in permutation:
            number_construction = number_construction * 10 + int(char)
        if isPrime(number_construction) == True:
            primeNumberList.append(number_construction)
    print(primeNumberList)

allPermutationList(fiveDigitsNumber)



```

```python

def countConsecutive(number: int, suma: int, n: int) -> bool:
    """
    The algorithm work with the following formula: n + (n + 1) + (n + 2)...
    This function will return a bool result for a number.
    That result represents if the number can be divided in a sum of consecutive numbers or not.
    
    Case 1: input is a odd(impar) number then we know 100% we cand make a sum a two consecutive numbers:
    int(sum / 2) + (int(sum/2)) + 1.Function return TRUE.
    Case 2: input number is a event(par) number then the function will search
    if the (number - sum) remainder is zero this is also True for every situation, 
    the function will repeat this step as the formula from the top, until sum gets greater than the input,
    in that case we know that no consecutive numbers can be sumed into the input, and the function return FALSE. 
    """
    if number % 2 != 0:
        return True
        
    if (number - suma) % n == 0:
        return True
        
    if suma > number:
        return False
        
    return countConsecutive(number, suma + n, n + 1)
  

number = 12
seDescompune = countConsecutive(12, 1, 2)

if seDescompune == True:
    print("Numarul: ", number, "se poate descompune in suma de numere consecutive.",seDescompune)
else:
    print("Numarul: ", number, "NU se poate descompune in suma de numere consecutive.",seDescompune)
```

(0.5 p) 1. Se dă un vector NumPy. Să se determine lista elementelor unice. Formați o matrice pătratică de dimensiune  `len(v)/2` care să conțină elementele unice din vector.
Completați cu cmmdc al elementelor unice pozițiile din matrice rămase libere.  

```python
import numpy as np
from enum import unique

def cmmdc (a:int, b:int) -> int:
    """
    This function will return the greatest common factor between 2 integer numbers.
    param a,b: the input numbers.
    """
    if b == 0:
        return a
    else:
        return cmmdc(b, a%b)
    
def cmmdcFromUniqueElements(unique_elements: int) -> int:
    """
    This function will return the greatest common factor from an list of unique elements using the cmmdc function.
    Function will iterate from second position of list to the last one and will put in the element variable the result.
    At the end the function will return a integer number that is the cmmdc for the all elements in the list.
    param unique_elements: a list of integer numbers.
    """
    element = unique_elements[0]
    for index in range(1, len(unique_elements)):
        element = cmmdc(element, unique_elements[index])
    return element

def generateMatrix(matrix: np.ndarray, vector_np: np.ndarray, size_np: int, unique_elements: int, size: int, cmmdc: int):
    counter = 0
    
    for index1 in range(0, (size_np // 2) + 1):
        for index2 in range(0, (size_np // 2) + 1):
            matrix[index1, index2] = cmmdc
                        
    for index1 in range(0, (size_np // 2) + 1):
        for index2 in range(0, (size_np // 2) + 1):            
            if counter <= size - 1:
                matrix[index1, index2] = unique_elements[counter]
                counter += 1
            else:
                break



vector_np = np.ndarray([10,5,15,20,15,15,10])
size_np = np.size(vector_np.shape)
print("size = ", size_np)

vector = [10,5,15,20,15,15,10]
unique_elements = np.unique(vector)
print("unique element vector: ", unique_elements)
size = len(unique_elements)
print("size unique = ", size)
cmmdc_unique_elements = cmmdcFromUniqueElements(unique_elements)
print("cmmdc din unique elements este: ", cmmdc_unique_elements)

matrix = np.zeros(shape = (int(size_np / 2) + 1, int(size_np / 2) + 1), dtype = int)
print("matricea de zero: \n", matrix)


    
generateMatrix(matrix, vector_np, size_np, unique_elements, size, cmmdc_unique_elements)
print("matricea finala: \n", matrix)
```
